```python
def assert_recursion(tab, i):
    """
    Assert that the array is in order.
    @param tab - the array
    @param i - the index of the current element
    @returns True if the tabulation is in order, False otherwise.
    """
    for i in range (0, i-2):
        # cas ou le tableau n'est pas trie de 0 à i-1
        if(tab[i] > tab[i+1]):
            return False
    return True

def assert_invariant_boucle(tab, k, i, x):
    """
    Assert that the invariant de boucle is respected.
    @param tab - the tab of the invariant de boucle
    @param k - the index of the first element of the boucle
    @param i - the index of the last element of the boucle
    @param x - the value of the invariant boucle
    @returns True if the invariant boucle is respected, False otherwise
    """
    # pour e de k à i tab[e] > x
    assertTeSupX=True
    #si tab[i-1] = tab[i]
    assertImoinsUnSupI=True
    # t[0..i-1] trié
    assertTableauTrie=True
    
    for e in range(k,i+1):
        if tab[e]<=x:
            assertTeSupX = False
    print("assertTeDupX =", assertTeSupX)

    if(tab[i-1]<=tab[i]):
        assertImoinsUnSupI = False
    print("assertImoinsUnSupI =", assertImoinsUnSupI)
    
    for j in range(0,i-2):
        if(tab[j]>tab[j+1]):
            assertTableauTrie = False
    print("assertTableauTrie =", assertTableauTrie)
    return (assertTeSupX or assertImoinsUnSupI) and assertTableauTrie

def assert_decalage(k,i,x,tab):
    """
    @param k - the pivot index
    @param i - the last index
    @param x - the x value
    @param tab - the tableau
    @returns True if the tableau is sorted and if x is greater than all the elements
    to its right. False otherwise.
    """
    assertTeSupX=True
    kPlus1Egalk = True
    assertTableauTrie=True
    for e in range(k,i+1):
        if tab[e]<=x:
            assertTeSupX = False
    print("assertTeDupX =", assertTeSupX)
    kPlus1Egalk = tab[k+1] == tab[k]    
    #tableau trié ?
    for j in range(0,i-2):
        if(tab[j]>tab[j+1]):
            assertTableauTrie = False
    print("assertTableauTrie =", assertTableauTrie)
    return (kPlus1Egalk and assertTeSupX and assertTableauTrie)

def assert_apres_dec_k(k,i,x,tab):
    """
    @param k - the index
    @param i - the last index
    @param x - the x value
    @param tab - the tableau
    @returns True if the tableau is sorted and if x is greater than all the elements
    to its right. False otherwise.
    """
    assertTeSupX=True
    kPlus1Egalk = True
    assertTableauTrie=True
    for e in range(k+1,i+1):
        if tab[e]<=x:
            assertTeSupX = False
    print("assertTeDupX =", assertTeSupX)
    kPlus1Egalk = tab[k+2] == tab[k+1]   
    print("kPlus1Egalk =", kPlus1Egalk)
    #tableau trié ?
    for j in range(0,i-1):
        if(tab[j]>tab[j+1]):
            assertTableauTrie = False
    print("assertTableauTrie =", assertTableauTrie)
    return (kPlus1Egalk and assertTeSupX and assertTableauTrie)

def assert_apres_fin_tant_que(k, i, x, tab):
    assertToutEltInfaX=True
    assertOinfX=True
    assertToutEltSupaX=True
    assertKPlusDeuxEgalKPlusUn=True
    assertTableauTrie=True
    assertTkInfOuEgalAx=tab[k]<=x or k<0 
    if(k>=0):
        for e in range(0,k+1):
            if(tab[e]>x):
                assertToutEltInfaX= False
        for o in range(k+1,i+1):
            if(tab[o]<=x):
                assertOinfX=False
        if(k==-1):
            for e in range(0,i+1):
                if(tab[e]<=x):
                    assertToutEltSupaX=False
    assertKPlusDeuxEgalKPlusUn= tab[k+2]==tab[k+1]
    for j in range(0,i-1):
        if(tab[j]<tab[j+1]):
            assertTableauTrie=False        
    return ((assertToutEltInfaX and assertOinfX or assertToutEltSupaX) and (assertKPlusDeuxEgalKPlusUn and assertTableauTrie)) or(assertTkInfOuEgalAx)

def assert_apres_insertion_x_dans_kplus1(tab, i):
    """
    Assert that the array is in order.
    @param tab - the array
    @param i - the index of the current element
    @returns True if the tabulation is in order, False otherwise.
    """
    for i in range (0, i-1):
        # cas ou le tableau n'est pas trie de 0 à i-1
        if(tab[i] > tab[i+1]):
            return False
    return True
```

```python


# tri par tas
def tri_tas(tab):
    """
    Tri par tas.
    @param tab - le tableau à trier
    @return le tableau trié
    """

    # construction du tas
    p = 0
    while p < len(tab)-1:
        ajouter_tas(tab, p, tab[p+1])
        p = p+1
    # fin tant que
    # tri a proprement parlé
    while p > 0:
        mini, p = retirer_min_tas(tab, p)
        tab[p + 1] = mini
        print(p)
    # fin tant que
    return tab

#implémenter predicat qui va remonter procédure qui 
# parcours tout l'arbre et lorqu'on arrive sur le
# fameux indice la trajectoire s'arrête
def ajouter_tas(tab, p, element):
    """
    Ajoute un element dans le tas.
    @param tab - le tableau à trier
    @param p - la position du tableau à trier
    @param element - l'element a ajouter
    """
    p = p + 1
    tab[p] = element
    i = p
    while i > 0 and tab[i//2] > tab[i]:
        tab[i], tab[i//2] = tab[i//2], tab[i]
        i = i//2


def retirer_min_tas(tab, p):
    """
    Retire le minimum dans le tas.
    @param tab - le tableau à trier
    @param p - la position du tableau à trier
    @return le minimum
    """
    mini = tab[0]
    tab[0] = tab[p]
    p = p-1
    i = 0
    fini = False
    while i <= p//2 and fini == False:
        if(p == 2*i or tab[2*i] < tab[2*i+1]):
            j = 2*i
        else:
            j = 2*i + 1
        # finsi
        if(tab[i] > tab[j]):
            tab[i], tab[j] = tab[j], tab[i]
            i = j
        else:
            fini = True
    return mini, p
```
